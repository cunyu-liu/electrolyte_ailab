```python
import pandas as pd
import re
import os

# 输入文件名
input_file = 'output_data13.csv'

# 自动生成输出文件名
base, ext = os.path.splitext(input_file)
output_file = f"collection1401_2032_clean.csv"

df = pd.read_csv(input_file)

main_columns = [
    'Electrolytes.Salts.Name',
    'Electrolytes.Solvents.Name',
    'Electrolytes.Additives.Name',
    'Anode.Name',
    'Cathode.Name'
]

target_columns = [
    'url',
    'Electrolytes.Salts.Name',
    'Electrolytes.Solvents.Name',
    'Electrolytes.Additives.Name',
    'Anode.Name',
    'Cathode.Name',
    'System of electrolyte.Type',
    'Properties of electrolyte.LUMO energy level.Value',
    'Properties of electrolyte.LUMO energy level.Unit',
    'Properties of electrolyte.HOMO energy level.Value',
    'Properties of electrolyte.HOMO energy level.Unit',
    'Properties of electrolyte.Viscosity (η).Value',
    'Properties of electrolyte.Viscosity (η).Unit',
    'Properties of electrolyte.Viscosity (η).Temperature.Value',
    'Properties of electrolyte.Viscosity (η).Temperature.Unit',
    'Properties of electrolyte.Dielectric constant (ε).Value',
    'Properties of electrolyte.Melting point.Value',
    'Properties of electrolyte.Melting point.Unit',
    'Properties of electrolyte.Boiling point.Value',
    'Properties of electrolyte.Boiling point.Unit',
    'Properties of electrolyte.Flash Point.Value',
    'Properties of electrolyte.Flash Point.Unit',
    'Performance.Energy density.Value',
    'Performance.Energy density.Unit',
    'Performance.Coulombic efficiency.Value',
    'Performance.Coulombic efficiency.Unit',
    'Performance.Cycle lifespan.Value',
    'Performance.Cycle lifespan.Unit',
    'Performance.Rate performance.Value',
    'Performance.Rate performance.Unit',
]

for main_col in main_columns:
    prefix = re.escape(main_col.replace('.Name', ''))
    pattern = re.compile(rf"^{prefix}(\d*)\.?Name$")
    matched_cols = [col for col in df.columns if pattern.match(col)]
    if not matched_cols:
        matched_cols = [main_col] if main_col in df.columns else []
    if matched_cols:
        df[main_col] = df[matched_cols].astype(str).apply(
            lambda x: ' & '.join(filter(lambda v: v.strip() and v.lower() != 'nan', x)), axis=1
        )
        to_drop = [c for c in matched_cols if c != main_col]
        if to_drop:
            df.drop(columns=to_drop, inplace=True)

final_df = df.reindex(columns=target_columns)
final_df.to_csv(output_file, index=False, encoding='utf-8-sig')
```
