```python
import os
import shutil
import pandas as pd

# 路径设置
source_folder = '/mnt/d/SEI_addictives/2008'
target_folder = '2008-v1'
csv_path = 'output_data_2008.csv'

# 读取CSV并提取url列，去除重复
df = pd.read_csv(csv_path)
url_set = set(df['url'].dropna().unique())

# 获取原始PDF文件列表
pdf_files = [f for f in os.listdir(source_folder) if f.endswith('.pdf')]
pdf_names = set(os.path.splitext(f)[0] for f in pdf_files)

# 确定哪些PDF没有被提取成功（即文件名不在url_set中）
pdf_not_extracted = [f for f in pdf_files if os.path.splitext(f)[0] not in url_set]

# 创建目标文件夹
os.makedirs(target_folder, exist_ok=True)

# 复制未提取的PDF文件
for pdf in pdf_not_extracted:
    shutil.copy(os.path.join(source_folder, pdf), os.path.join(target_folder, pdf))

# 输出统计信息
print(f"原文件夹 PDF 数目: {len(pdf_files)}")
print(f"CSV 中不重复的 url 数目: {len(url_set)}")
print(f"新文件夹中 PDF 数目（未提取成功的）: {len(pdf_not_extracted)}")
```
